```python
""" 
Simple GAN using fully connected layers

Programmed by Aladdin Persson <aladdin.persson at hotmail dot com>
https://github.com/aladdinpersson/Machine-Learning-Collection/blob/master/ML/Pytorch/GANs/1.%20SimpleGAN/fc_gan.py

* 2020-11-01: Initial coding
* 2022-12-20: Small revision of code, checked that it works with latest PyTorch version
"""
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.datasets as datasets
from torch.utils.data import DataLoader, random_split

import torchvision.transforms as transforms
```

```python
# Check if CUDA is available
is_cuda_available = torch.cuda.is_available()
print("Is CUDA available:", is_cuda_available)

# Determine the device to use: GPU (CUDA), Apple Silicon (MPS), or CPU
DEVICE = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print("Using device:", DEVICE)
```

```python
class Discriminator(nn.Module):
    def __init__(self, in_features):
        super().__init__()
        self.disc = nn.Sequential(
            nn.Linear(in_features, 128),
            nn.LeakyReLU(0.01),
            nn.Linear(128, 1),
            nn.Sigmoid(),
        )
        
    def forward(self, x):
        return self.disc(x)
```

```python
class Generator(nn.Module):
    def __init__(self, z_dim, img_dim):
        super().__init__()
        self.gen = nn.Sequential(
            nn.Linear(z_dim, 256),
            nn.LeakyReLU(0.01),
            nn.Linear(256, img_dim),
            # nn.Tanh(),  # originally this code uses Tanh to normalize inputs to [-1, 1], and a normalise preprocessing step to get MNIST to be [-1,1]
            nn.Sigmoid(), # But I don't get it. MNIST is originally [0,1], so why not just not just leave it and use a sigmoid?
        )

    def forward(self, x):
        return self.gen(x)
```

```python
# Hyperparameters etc.
device = "cuda" if torch.cuda.is_available() else "cpu"
lr = 3e-4
z_dim = 64
image_dim = 28 * 28 * 1  # 784, MNIST is 28x28x1
batch_size = 32
num_epochs = 100

disc = Discriminator(image_dim).to(device)
gen = Generator(z_dim, image_dim).to(device)
fixed_noise = torch.randn((batch_size, z_dim)).to(device)
opt_disc = optim.Adam(disc.parameters(), lr=lr)
opt_gen = optim.Adam(gen.parameters(), lr=lr)
criterion = nn.BCELoss()
```

```python
transform = transforms.Compose([transforms.ToTensor()])

# Download the MNIST dataset
mnist_trainset = datasets.MNIST(root='~/.pytorch/MNIST_data/', train=True, download=True, transform=transform)

# Splitting the dataset into train and validation sets
train_size = int(0.8 * len(mnist_trainset))
validation_size = len(mnist_trainset) - train_size
train_dataset, validation_dataset = random_split(mnist_trainset, [train_size, validation_size])

# Download and load the test data
test_dataset = datasets.MNIST(root='~/.pytorch/MNIST_data/', train=False, download=True, transform=transform)

trainloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, drop_last=False)
validationloader = DataLoader(validation_dataset, batch_size=batch_size, shuffle=True, drop_last=False)
testloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True, drop_last=False)
```

Training

```python
save_path = f'MNIST_gan_{num_epochs}_epoch.pth'

for epoch in range(num_epochs):
    for batch_idx, (real, _) in enumerate(trainloader):
        real = real.view(-1, 784).to(device)
        batch_size = real.shape[0]

        ### Train Discriminator: max log(D(x)) + log(1 - D(G(z)))
        noise = torch.randn(batch_size, z_dim).to(device)
        fake = gen(noise)
        disc_real = disc(real).view(-1)
        lossD_real = criterion(disc_real, torch.ones_like(disc_real))
        disc_fake = disc(fake).view(-1)
        lossD_fake = criterion(disc_fake, torch.zeros_like(disc_fake))
        lossD = (lossD_real + lossD_fake) / 2
        disc.zero_grad()
        lossD.backward(retain_graph=True)
        opt_disc.step()

        ### Train Generator: min log(1 - D(G(z))) <-> max log(D(G(z))
        # where the second option of maximizing doesn't suffer from
        # saturating gradients
        output = disc(fake).view(-1)
        lossG = criterion(output, torch.ones_like(output))
        gen.zero_grad()
        lossG.backward()
        opt_gen.step()

        if batch_idx == 0:
            print(
                f"Epoch [{epoch}/{num_epochs}] Batch {batch_idx}/{len(trainloader)} \
                      Loss D: {lossD:.4f}, loss G: {lossG:.4f}"
            )

            with torch.no_grad():
                fake = gen(fixed_noise).reshape(-1, 1, 28, 28)
                data = real.reshape(-1, 1, 28, 28)
                img_grid_fake = torchvision.utils.make_grid(fake, normalize=True)
                img_grid_real = torchvision.utils.make_grid(data, normalize=True)

                # Display fake images
                plt.figure(figsize=(10,10))
                plt.imshow(img_grid_fake.permute(1, 2, 0).cpu().numpy())
                plt.title("Generated Images")
                plt.show()
                
                # Optionally display real images
                plt.figure(figsize=(10,10))
                plt.imshow(img_grid_real.permute(1, 2, 0).cpu().numpy())
                plt.title("Real Images")
                plt.show()

            torch.save({
                'epoch': epoch,
                'generator_state_dict': gen.state_dict(),
                'discriminator_state_dict': disc.state_dict(),
                'optimizer_gen_state_dict': opt_gen.state_dict(),
                'optimizer_disc_state_dict': opt_disc.state_dict(),
            }, save_path)

print(f"Final model saved at {save_path} from epoch {num_epochs}")
```

```python
import torch
import matplotlib.pyplot as plt
import torchvision

# Load the saved model
# save_path = f'MNIST_gan_{num_epochs}_epoch.pth'
save_path = 'MNIST_gan_100_epoch.pth'

checkpoint = torch.load(save_path)
gen.load_state_dict(checkpoint['generator_state_dict'])
epoch = checkpoint['epoch']
print(f'Trained on {epoch} epochs')

gen.eval()  # Set to evaluation mode
```

```python
# Generate images to display
fixed_noise = torch.randn((batch_size, z_dim)).to(device)

with torch.no_grad():
    fake_images = gen(fixed_noise).reshape(-1, 1, 28, 28)
    img_grid = torchvision.utils.make_grid(fake_images, normalize=True)

# Display the images
plt.figure(figsize=(10,10))
plt.imshow(img_grid.permute(1, 2, 0).cpu().numpy())
plt.title("Generated Images")
plt.show()
```

```python
# Generate images to save
num_images_to_generate = 100  # The total number of images you want to generate
generated_images = []  # List to hold the generated images

# Generate images in batches of 32
while len(generated_images) < num_images_to_generate:
    fixed_noise = torch.randn((batch_size, z_dim)).to(device)
    with torch.no_grad():
        images = gen(fixed_noise).reshape(-1, 1, 28, 28)
        generated_images.extend(images.cpu().numpy())  # Add generated images to the list

# Trim the list to only have 100 images
generated_images = generated_images[:num_images_to_generate]

# Convert list to a tensor
generated_images_tensor = torch.tensor(generated_images)

# Save the tensor of 100 generated images
torch.save(generated_images_tensor, 'gan_generated_mnist_images.pt')
```
